## Clustering - k-means

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
features, true_labels = make_blobs(
    n_samples=200,
    centers=3,
    cluster_std=2.75,
    random_state=42
)
```

```python
plt.scatter([features[i][0] for i in range(len(features))], [features[i][1] for i in range(len(features))])
```

```python
features[1][1]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
)
kmeans.fit(scaled_features)
```

```python
centroids = kmeans.cluster_centers_
```

```python
plt.scatter([scaled_features[i][0] for i in range(len(scaled_features))], [scaled_features[i][1] for i in range(len(scaled_features))])
plt.scatter([centroids[i][0] for i in range(len(centroids))], [centroids[i][1] for i in range(len(centroids))], c='r')
```

```python
print(f'Iterations required to convegre: {kmeans.n_iter_}')
```
